```python
import pandas as pd
import random

# -----------------------------
# 1. Sample Data Lists
# -----------------------------
event_names = [
    "AI Workshop", "Cloud Computing Seminar", "Coding Competition",
    "Cultural Fest", "Cybersecurity Workshop", "Hackathon",
    "Data Science Seminar", "Robotics Demo", "Startup Pitch Event"
]

departments = [
    "Computer Science", "Electronics", "Mechanical",
    "Civil", "IT", "Management"
]

event_types = ["Workshop", "Seminar", "Competition", "Technical Talk"]

sample_comments = [
    "Great event, very informative!",
    "Could be more interactive.",
    "Loved the speaker and content.",
    "The event started late.",
    "Amazing experience!",
    "Too many participants, difficult to manage.",
    "Well organized and detailed.",
    "Not satisfied with the arrangements.",
    "Excellent session, learned a lot!",
    "Average event, could improve the pace."
]

# -----------------------------
# 2. Generate 200 sample rows
# -----------------------------
rows = []

for i in range(200):
    row = {
        "Student ID": random.randint(100, 999),
        "Event Name": random.choice(event_names),
        "Department": random.choice(departments),
        "Event Type": random.choice(event_types),
        "Overall Rating (1-10)": random.randint(1, 10),
        "Comments": random.choice(sample_comments)
    }
    rows.append(row)

# -----------------------------
# 3. Convert to DataFrame
# -----------------------------
df = pd.DataFrame(rows)

# -----------------------------
# 4. Save to CSV
# -----------------------------
df.to_csv("event_feedback.csv", index=False)

df.head()
```

## Rename the columns

```python
df = pd.read_csv(r"C:\Users\Administrator\Documents\DS_03\data\Raw.csv")

df = df.rename(columns={
    'Student ID': 'student_id',
    'Well versed with the subject': 'well_versed',
    'Explains concepts in an understandable way': 'explains_concepts',
    'Use of presentations': 'presentations',
    'Degree of difficulty of assignments': 'assignment_difficulty',
    'Solves doubts willingly': 'solves_doubts',
    'Structuring of the course': 'course_structure',
    'Provides support for students going above and beyond': 'student_support',
    'Course recommendation based on relevance': 'recommendation'
})
```

## Rename Columns

```python
df.columns = df.columns.str.lower().str.replace(" ", "_")
```

## Handle Missing Comments

```python
df['comments'] = df['comments'].fillna("")
```

```python
df = df.drop(columns=['Unnamed: 0'], errors='ignore')
```

## Convert Rating Columns to Numeric

```python
rating_cols = [
    'well_versed',
    'explains_concepts',
    'presentations',
    'assignment_difficulty',
    'solves_doubts',
    'course_structure',
    'student_support',
    'recommendation'
]

for col in rating_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```
