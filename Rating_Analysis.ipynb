```python
import pandas as pd
```

## Load cleaned data

```python
df = pd.read_csv(r"C:\Users\Administrator\Documents\DS_03\outputs\event_feedback.csv")
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.histplot(df['Overall_Rating'], kde=True)
plt.title("Distribution of Overall Ratings")
plt.show()
```

## Satisfaction by Event Type

```python
sns.boxplot(x='Event Type', y='Overall_Rating', data=df)
plt.title("Event Type vs Overall Rating")
```

## Top 3 Events

```python
df.groupby('Event Name')['Overall_Rating'].mean().sort_values(ascending=False).head(3)
```

## Department-wise Avg Rating

```python
dept_avg = df.groupby('Department')['Overall_Rating'].mean().sort_values(ascending=False)
dept_avg
```

## Correlation Heatmap (Ratings Only)

```python
num_df = df.select_dtypes(include='number')

sns.heatmap(num_df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Between Rating Parameters")
plt.show()
```
